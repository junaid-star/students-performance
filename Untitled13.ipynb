```python
import seaborn as sns
data['total']=(data['math score']+data['reading score']+data['writing score'])/3
h = data["total"].head(20)
plt.plot(h)
```

```python
passmark=35
data['Reading_PassStatus'] = np.where(data['reading score']<passmark, 'F', 'P')
data.Reading_PassStatus.value_counts()
```

```python
sns.countplot(x='parental level of education', data = data, hue='Reading_PassStatus', palette='dark')
plt.show()
```

```python
passmarks=33
data['pass_reading'] = np.where(data['reading score']< passmarks, 'Fail', 'Pass')
data['pass_reading'].value_counts(dropna = False).plot.pie(colors = ['pink', 'yellow'])

plt.title('Pass/Fail in Reading', fontweight = 30, fontsize = 20)
plt.xlabel('status')
plt.ylabel('count')
plt.show()
```

```python
course_gender = data.groupby(['gender','test preparation course']).mean().reset_index()
```

```python
sns.factorplot(x='gender', y='total', hue='test preparation course', data=course_gender, kind='bar')
```
